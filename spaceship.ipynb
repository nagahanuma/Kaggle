```python
df=pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
df.head(5)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#car_data.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True) 
tr_x=df[["PassengerId","HomePlanet","CryoSleep","Destination","Age","VIP",]]
tr_y=df[["Transported"]]
```

```python
tr_x.isnull().sum()
```

```python
tr_x['CryoSleep']=tr_x['CryoSleep'].fillna(tr_x['CryoSleep'].mode()[0])
tr_x['VIP']=tr_x['VIP'].fillna(tr_x['VIP'].mode()[0])
tr_x['Destination']=tr_x['Destination'].fillna(tr_x['Destination'].mode()[0])
tr_x['HomePlanet']=tr_x['HomePlanet'].fillna(tr_x['HomePlanet'].mode()[0])
tr_x['Age']=tr_x['Age'].fillna(tr_x['Age'].mean())
```

```python
#tx=pd.get_dummies(tr_x, columns = ['Destination'])
#df['salary'] = df['salary'].fillna(df['salary'].mode()[0])

tr_x.replace({'HomePlanet':{'Earth':0,'Europa':1,'Mars':2}},inplace=True)
#tr_x.replace({'VIP':{'False':0,'True':1}},inplace= True)
#test_x["Sex"].replace("male",1,inplace=True)
tr_x['VIP']=tr_x['VIP'].astype(int)
tr_x['CryoSleep']=tr_x['CryoSleep'].astype(int)
#tr_x.replace({'CryoSleep':{'False':0,'True':1}},inplace= True)
tr_x.replace({'Destination':{'55 Cancri e':0,'PSO J318.5-22':1,'TRAPPIST-1e':2}},inplace=True)
passengerId=[]
tr_x

```

```python
x=tr_x['PassengerId']
pid=[]
for i in range(0,len(x)):
    p=x[i]
    ne=(p[0:4]+p[5:7])
    une=int(ne)
    pid.append(une)
tr_x['PassengerId']=pid
tr_x
    
```

```python
t_x,cv_x,t_y,cv_y=train_test_split(tr_x,tr_y,test_size=0.25)
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(tr_x,tr_y)
```

```python
acc=rf.score(cv_x,cv_y)
print(acc*100)
```
