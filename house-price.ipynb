```python
train_y=df[["SalePrice"]]
train_y.head()
```

```python
from sklearn.model_selection import train_test_split
tr_x,cv_x,tr_y,cv_y=train_test_split(train_x,train_y,test_size=0.25)
print(tr_x)
print(tr_y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
classifier=rf.fit(tr_x,tr_y)
y_pred=rf.predict(cv_x)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(cv_y,y_pred)
print(r2*100)
#acc=rf.score(cv_x,cv_y)
#print(acc*100)
```
